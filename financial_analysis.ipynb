```python
import pandas as pd 
```

```python
df=pd.read_csv("financials_companies.csv")
```

```python
df.head()
```

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue (USD millions)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income (USD millions)'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets (USD millions)'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities (USD millions)'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow from Operating Activities (USD millions)'].pct_change() * 100
```

```python
df
```

```python
import matplotlib.pyplot as plt

for company in df['Company'].unique():
    subset = df[df['Company'] == company]
    plt.figure()
    plt.plot(subset['Fiscal Year'], subset['Total Revenue (USD millions)'], marker='o')
    plt.title(f"{company} - Revenue Trend")
    plt.xlabel('Fiscal Year')
    plt.ylabel('Revenue (USD millions)')
    plt.show()
```

### Key Findings
- Microsoft shows strong YoY revenue and cash flow growth in FY2024.
- Teslas 2024 net income dropped despite revenue stability.
- Apples revenue plateaued, but cash flow remained robust.
